```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

```python
data=pd.read_csv("Automobile_data.csv")
```

### to change dtype of any coloumn in csv file

```python
data['horsepower']=pd.to_numeric(data['horsepower'],errors='coerce').fillna(0).astype(int)
data['normalized-losses']=pd.to_numeric(data['normalized-losses'],errors='coerce').fillna(0).astype(int)
print(data.dtypes)
```

```python
data.head()
```

```python
sns.distplot(data['engine-size'])
plt.show()
```

```python
sns.distplot(data['engine-size'],kde=False, rug=True)
plt.show()
```

```python
sns.jointplot(data["engine-size"],data["horsepower"])
plt.show()
```

```python
sns.jointplot(data["engine-size"],data["horsepower"],kind='reg')
plt.show()
```

```python
sns.jointplot(data["engine-size"],data["horsepower"],kind='hex')
plt.show()
```

```python
sns.jointplot(data["engine-size"],data["horsepower"],kind='kde')
plt.show()
```

```python
sns.pairplot(data[["normalized-losses",'horsepower','engine-size']])
```

```python
sns.stripplot(data['fuel-type'],data['horsepower'])
```

```python
sns.swarmplot(data['fuel-type'],data['horsepower'])
```

```python
sns.boxplot(data['num-of-doors'],data['horsepower'],hue=data['fuel-type'])
```

```python
sns.barplot(data['body-style'],data['horsepower'],hue=data['engine-location'])
```

```python
sns.countplot(data['body-style'])
```

```python
sns.pointplot(data['fuel-system'],data['horsepower'], hue=data["num-of-doors"])
```

```python
sns.factorplot(x='fuel-type',
               y='horsepower',
              hue='num-of-doors',
              col='engine-location',
              data=data,
              kind='swarm')
```
